```python
prediction_data = test_data.drop(['PassengerId','Name','Ticket','Embarked','Cabin'],axis=1)
dum = pd.get_dummies(prediction_data.Sex)
kaggle = pd.concat([prediction_data,dum],axis=1)
kaggle.drop(['Sex','male'],axis=1,inplace=True)
kaggle['Age'].fillna(kaggle['Age'].mean(),inplace=True)
kaggle['Fare'].fillna(kaggle['Fare'].mean(), inplace=True)
```
